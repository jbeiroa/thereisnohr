```python
def extract_markdown(path: Path) -> str:
    if not path.exists():
        raise FileNotFoundError(f"Resume not found: {path}")
    doc = pymupdf.open(path)
    return pymupdf4llm.to_markdown(doc, show_progress=False)
```
