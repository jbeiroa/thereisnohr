# Extraction Service Testing

This notebook tests the `ExtractionService` by extracting structured data from job descriptions and parsed resume sections.

```python
import sys
from pathlib import Path

# Ensure project root is in sys.path
project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.extract.service import ExtractionService
from src.llm.factory import build_default_llm_client
from notebooks._utils import print_checkpoint
```

```python
client = build_default_llm_client()
service = ExtractionService(llm_client=client)
print_checkpoint("ExtractionService initialized with default LLM client")
```

## 1. Job Requirements Extraction

Extract structured requirements from a sample job description.

```python
job_desc = """
Senior Python Engineer
We are looking for a Senior Python Engineer with 5+ years of experience in backend development.
Requirements:
- Strong proficiency in Python and FastAPI.
- Experience with PostgreSQL and vector databases (pgvector).
- Familiarity with LLM orchestration (LangChain, Metaflow).
- BS or MS in Computer Science or a related field.
- AWS certifications are a plus.
"""

job_requirements = service.extract_job_requirements(job_desc)
print_checkpoint("Job requirements extracted")
print(job_requirements.model_dump_json(indent=2))
```

## 2. Candidate Signals Extraction

Extract structured signals from sample resume sections.

```python
sections = {
    "summary": "Experienced Software Engineer with a focus on AI/ML and backend systems.",
    "experience": "Senior Developer at Tech Corp (2020-Present). Built scalable microservices using Python and Go. Led a team of 5 engineers.",
    "education": "BS in Computer Engineering, MIT (2016-2020).",
    "skills": "Python, Go, Kubernetes, Docker, PyTorch."
}

signals, diagnostics = service.extract_with_diagnostics(sections)
print_checkpoint("Candidate signals extracted with diagnostics")

print("--- Signals ---")
print(signals.model_dump_json(indent=2))

print("\n--- Diagnostics ---")
print(diagnostics.model_dump_json(indent=2))
```
